```python
import json, urllib.request
d = json.loads(urllib.request.urlopen("https://emojidome.xkcd.com/2131/socket ").read().decode('utf-8'))
for g in d['bracket']['played'][0]:
    c1, c2 = g['game']
    print(f"{c1['score']:>5d} {c1['competitor']}-{c2['competitor']} {c2['score']:5d}", end='')
    s1, s2 = c1['score'], c2['score']
    print(f"  {s1/(s1+s2)*100:.1f} - {s2/(s1+s2)*100:.1f}")
```
